```kotlin
%use dataframe
```

```kotlin
var df = DataFrame.readCSV("train.csv", delimiter='\t')
```

```kotlin
df = df.dropNA { tokens }
df.head()
```

```kotlin
%use coroutines
```

```kotlin
import java.util.concurrent.ConcurrentHashMap

val marks = List(2) { ConcurrentHashMap<String, Int>() }
```

```kotlin
import java.util.concurrent.atomic.AtomicInteger

val i = AtomicInteger()
```

```kotlin
val size = 1514340

runBlocking {
    df.rows().forEach { row ->
        launch {
            row.tokens!!.split(" ").windowed(2, step=2).forEach {
                val word = it[0]
                if (!marks[row.DEF].containsKey(word)) {
                    marks[row.DEF][word] = 0
                }
                marks[row.DEF][word] = marks[row.DEF][word]!!.plus(it[1].toInt())     
            }
            val iter = i.incrementAndGet()
            if (iter % 5000 == 0) println("$iter ${iter.toDouble() / size}")
        }
    }
}
```

```kotlin
val freq = HashMap<String, Double>()
val total_cnt = HashMap<String, Int>()

val it = marks[0].keys + marks[1].keys
it.forEach { word ->
    val total = marks[0].getOrDefault(word, 0) + marks[1].getOrDefault(word, 0)
    total_cnt[word] = total
    if (total >= 100000) {
        freq[word] = marks[1].getOrDefault(word, 0).toDouble() / total 
    }
}
```

```kotlin
freq.keys.sortedBy { freq[it]!! }.take(100).forEach {
    println("$it ${freq[it]} ${total_cnt[it]}")
}
```

```kotlin
File("outputGraph.csv").printWriter(charset = charset("Windows-1251")).use { writer ->
    total_cnt.forEach {
        writer.print("${it.key};${it.value}\r\n")
    }
}
```
